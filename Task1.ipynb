```python
df_c = df.copy()
```

```python
df_c = df_c.dropna(subset = ['Full Name', 'Department', 'Job Title'])
```

```python
df_c['Age'] = df_c['Age'].fillna(df_c['Age'].mean())
```

```python
df_c.isna().sum()
```

```python
df_c['Annual Salary'] = df_c['Annual Salary'].fillna(df_c['Annual Salary'].mean())
```

```python
df_c.isna().sum()
```

```python
df_c['Ethnicity'] = df_c['Ethnicity'].fillna(df_c['Ethnicity'].mode()[0])
```

```python
df_c['Country'] = df_c['Country'].fillna(df_c['Country'].mode()[0])
```

```python
df_c['City'] = df_c['City'].fillna(df_c['City'].mode()[0])
```

```python
df_c['Bonus %'] = df_c['Bonus %'].fillna(df_c['Bonus %'].mean())
```

```python
df_c.isna().sum()
```

```python
df_c['Hire Date'] = df_c['Hire Date'].fillna(0000-00-00)
```

```python
highest_salary = df_c[df_c['Annual Salary'] == df_c['Annual Salary'].max()]
print(f"{highest_salary}")
```

```python
#dept_group = df_c.groupby('Department')[['Age', 'Annual Salary']].mean()
dept_group = df_c.groupby('Department').agg({
    'Age' : 'mean',
    'Annual Salary': 'mean'
})

print(dept_group)
```

```python
group = df_c.groupby(['Department', 'Ethnicity']).agg({
    'Age' : ['max', 'min'],
    'Annual Salary': 'median'
})
print(group)
```

```python
df_c.groupby('Department').agg({
    'Annual Salary': lambda x: x.max() - x.min()  # salary range
})
```

```python
# Flatten MultiIndex columns
group.columns = ['Max Age', 'Min Age', 'Median Salary']
group = group.reset_index()
```

```python
group
```

```python
with pd.ExcelWriter("employee_summary.xlsx") as writer:
    df_c.to_excel(writer, sheet_name="Cleaned Data", index=False)
    group.to_excel(writer, sheet_name="Grouped Summary", index=False)
```
